# FEATURE SCALING

```python
import numpy as np
```

```python
test_data = pd.DataFrame({'x':np.random.randint(-500,100,1000),'y':np.random.randint(50,1000,1000),'z':np.random.randint(-500,100,1000)})
```

```python
test_data
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(test_data.x,test_data.y,test_data.z)
plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
nm=Normalizer()
```

```python
normaldata=nm.fit_transform(test_data)
```

```python
normaldata
```

```python
normal_df=pd.DataFrame(normaldata,columns=['a','b','c'])
```

```python
normal_df
```

```python
pl=plt.axes(projection='3d')
pl.scatter3D(normal_df.a,normal_df.b,normal_df.c)
plt.figure()
```

# MISSING VALUES

```python
data={'a':[1,2,3,4,5],'b':[6,7,np.nan,9,0],'c':[np.nan,22,np.nan,44,55]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
from sklearn.impute import SimpleImputer 
```

```python
si=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
si.fit(df)
```

```python
si.transform(df)
```

```python
sm=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
sm.fit(df)
```

```python
sm.transform(df)
```

```python
smm=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
smm.fit(df)
```

```python
smm.transform(df)
```
